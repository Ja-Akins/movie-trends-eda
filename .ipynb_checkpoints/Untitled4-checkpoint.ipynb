```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast

# Set style for better visualizations
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Display settings
pd.set_option('display.max_columns', None)
```

# 1. Data Loading and Merging

```python
# Load datasets
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")

# Merge datasets
merged_data = movies.merge(credits, left_on="id", right_on="movie_id", how="inner")
merged_data.set_index('id', inplace=True)

# Select relevant columns
columns_needed = [
    'budget', 'genres', 'keywords', 'original_title', 'overview', 
    'popularity', 'production_companies', 'release_date', 'revenue', 
    'runtime', 'status', 'tagline', 'vote_average', 'vote_count', 
    'cast', 'crew'
]
merged_data = merged_data[columns_needed]

print(f"Initial dataset shape: {merged_data.shape}")
```

# 2. Data Cleaning & Preparation

```python
def clean_and_prepare_data(df):
    """Clean and prepare the movie dataset"""
    
    # Create a copy to avoid modifying original
    df_clean = df.copy()
    
    # Filter out unrealistic budget/revenue values
    df_clean = df_clean[(df_clean['budget'] > 1000) & (df_clean['revenue'] > 1000)]
    
    # Handle missing values in critical columns
    df_clean = df_clean.dropna(subset=['budget', 'revenue', 'runtime'])
    
    # Convert release_date to datetime and extract year
    df_clean['release_date'] = pd.to_datetime(df_clean['release_date'], errors='coerce')
    df_clean['release_year'] = df_clean['release_date'].dt.year
    df_clean = df_clean.dropna(subset=['release_year'])
    
    # Process genres from string to list
    df_clean['genres'] = df_clean['genres'].apply(ast.literal_eval)
    df_clean['genres'] = df_clean['genres'].apply(
        lambda x: [d['name'] for d in x] if isinstance(x, list) else []
    )
    df_clean['genre_count'] = df_clean['genres'].apply(len)
    
    # Create financial metrics
    df_clean['profit'] = df_clean['revenue'] - df_clean['budget']
    df_clean['ROI'] = df_clean['revenue'] / df_clean['budget'].replace(0, pd.NA)
    
    return df_clean

# Apply cleaning function
cleaned_data = clean_and_prepare_data(merged_data)
print(f"Cleaned dataset shape: {cleaned_data.shape}")
print(f"Missing values:\n{cleaned_data.isnull().sum()}")
```

# 3. Exploratory Data Analysis

## 3.1 Descriptive Statistics

```python
def get_financial_summary(df):
    """Generate financial summary statistics"""
    financial_cols = ['budget', 'revenue', 'profit', 'ROI', 'runtime']
    summary = df[financial_cols].agg(['count', 'mean', 'median', 'std', 'min', 'max'])
    
    # Format large numbers for better readability
    for col in ['budget', 'revenue', 'profit']:
        summary[col] = summary[col].apply(lambda x: f"{x:,.0f}")
    
    return summary

# Display financial summary
financial_summary = get_financial_summary(cleaned_data)
print("Financial Summary:")
display(financial_summary)

# Key metrics comparison
print("\nKey Financial Metrics:")
print(f"Median budget: ${cleaned_data['budget'].median():,.0f}")
print(f"Median revenue: ${cleaned_data['revenue'].median():,.0f}")
print(f"Average budget: ${cleaned_data['budget'].mean():,.0f}")
print(f"Average revenue: ${cleaned_data['revenue'].mean():,.0f}")
```

## 3.2 Genre Analysis

```python
def analyze_genres(df):
    """Analyze movie genres and their financial performance"""
    
    # Count movies per genre
    genre_counts = df['genres'].explode().value_counts()
    
    # Financial analysis by genre
    genre_financials = (
        df.assign(genre=df['genres'])
        .explode('genres')
        .groupby('genres')
        .agg({
            'original_title': 'count',
            'revenue': ['mean', 'median'],
            'profit': ['mean', 'median'],
            'ROI': 'median'
        })
        .round(2)
    )
    
    # Flatten column names
    genre_financials.columns = ['count', 'avg_revenue', 'median_revenue', 
                               'avg_profit', 'median_profit', 'median_ROI']
    genre_financials = genre_financials.sort_values('count', ascending=False)
    
    return genre_counts, genre_financials

# Get genre analysis
genre_counts, genre_financials = analyze_genres(cleaned_data)

print("Top 10 Genres by Movie Count:")
print(genre_counts.head(10))
print("\nFinancial Performance by Genre:")
print(genre_financials.head(10))
```

## 3.3 Visualizations

```python
def create_genre_visualizations(genre_counts, genre_financials):
    """Create visualizations for genre analysis"""
    
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
    # Plot 1: Most common genres
    genre_counts.head(10).plot(kind='bar', ax=axes[0], color='skyblue')
    axes[0].set_title("Top 10 Genres by Movie Count", fontsize=14, fontweight='bold')
    axes[0].set_ylabel("Number of Movies", fontsize=12)
    axes[0].tick_params(axis='x', rotation=45)
    
    # Plot 2: Average revenue by genre
    genre_financials.head(10)['avg_revenue'].sort_values(ascending=True).plot(
        kind='barh', ax=axes[1], color='lightcoral'
    )
    axes[1].set_title("Average Revenue by Genre (Top 10)", fontsize=14, fontweight='bold')
    axes[1].set_xlabel("Average Revenue ($)", fontsize=12)
    
    plt.tight_layout()
    plt.show()

# Create visualizations
create_genre_visualizations(genre_counts, genre_financials)
```

## 3.4 Additional Analysis: Yearly Trends

```python
def analyze_yearly_trends(df):
    """Analyze movie trends over years"""
    
    yearly_stats = df.groupby('release_year').agg({
        'original_title': 'count',
        'revenue': 'mean',
        'budget': 'mean',
        'profit': 'mean',
        'vote_average': 'mean'
    }).round(2)
    
    yearly_stats.columns = ['movie_count', 'avg_revenue', 'avg_budget', 
                           'avg_profit', 'avg_rating']
    
    return yearly_stats

# Get yearly trends
yearly_trends = analyze_yearly_trends(cleaned_data)

print("Recent Yearly Trends (Last 5 years):")
print(yearly_trends.tail().sort_index(ascending=False))

# Plot yearly trends
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Movie count over years
axes[0,0].plot(yearly_trends.index, yearly_trends['movie_count'], marker='o')
axes[0,0].set_title("Number of Movies per Year")
axes[0,0].set_ylabel("Movie Count")

# Average revenue over years
axes[0,1].plot(yearly_trends.index, yearly_trends['avg_revenue'], marker='o', color='green')
axes[0,1].set_title("Average Revenue per Year")
axes[0,1].set_ylabel("Average Revenue ($)")

# Average budget over years
axes[1,0].plot(yearly_trends.index, yearly_trends['avg_budget'], marker='o', color='orange')
axes[1,0].set_title("Average Budget per Year")
axes[1,0].set_ylabel("Average Budget ($)")

# Average rating over years
axes[1,1].plot(yearly_trends.index, yearly_trends['avg_rating'], marker='o', color='red')
axes[1,1].set_title("Average Rating per Year")
axes[1,1].set_ylabel("Average Rating")

plt.tight_layout()
plt.show()
```

## 3.5 Correlation Analysis

```python
def analyze_correlations(df):
    """Analyze correlations between different metrics"""
    
    correlation_cols = ['budget', 'revenue', 'profit', 'popularity', 
                       'vote_average', 'vote_count', 'runtime']
    
    correlation_matrix = df[correlation_cols].corr()
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
                square=True, fmt='.2f')
    plt.title("Correlation Matrix of Movie Metrics", fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    return correlation_matrix

# Get correlation analysis
correlation_matrix = analyze_correlations(cleaned_data)
print("Strongest Correlations:")
print(f"Revenue vs Profit: {correlation_matrix.loc['revenue', 'profit']:.3f}")
print(f"Popularity vs Revenue: {correlation_matrix.loc['popularity', 'revenue']:.3f}")
```

# 4. Summary and Key Insights

```python
def generate_insights(df, genre_financials):
    """Generate key insights from the analysis"""
    
    print("="*50)
    print("KEY INSIGHTS FROM MOVIE DATA ANALYSIS")
    print("="*50)
    
    # Top performing genres
    top_revenue_genre = genre_financials.loc[genre_financials['avg_revenue'].idxmax()]
    top_profit_genre = genre_financials.loc[genre_financials['avg_profit'].idxmax()]
    top_roi_genre = genre_financials.loc[genre_financials['median_ROI'].idxmax()]
    
    print(f"\n1. Most Profitable Genres:")
    print(f"   - Highest Average Revenue: {top_revenue_genre.name} (${top_revenue_genre['avg_revenue']:,.0f})")
    print(f"   - Highest Average Profit: {top_profit_genre.name} (${top_profit_genre['avg_profit']:,.0f})")
    print(f"   - Best ROI: {top_roi_genre.name} ({top_roi_genre['median_ROI']:.1f}x)")
    
    # Financial success rate
    successful_movies = len(df[df['profit'] > 0])
    success_rate = (successful_movies / len(df)) * 100
    
    print(f"\n2. Financial Success Rate: {success_rate:.1f}% of movies are profitable")
    
    # Recent trends
    recent_years = df[df['release_year'] >= 2010]
    avg_budget_recent = recent_years['budget'].mean()
    avg_revenue_recent = recent_years['revenue'].mean()
    
    print(f"\n3. Recent Trends (2010+):")
    print(f"   - Average Budget: ${avg_budget_recent:,.0f}")
    print(f"   - Average Revenue: ${avg_revenue_recent:,.0f}")
    
    # Most common genre combinations
    df['genre_combination'] = df['genres'].apply(lambda x: ', '.join(sorted(x)))
    common_combinations = df['genre_combination'].value_counts().head(5)
    
    print(f"\n4. Most Common Genre Combinations:")
    for i, (combo, count) in enumerate(common_combinations.items(), 1):
        print(f"   {i}. {combo} ({count} movies)")

# Generate insights
generate_insights(cleaned_data, genre_financials)
```
