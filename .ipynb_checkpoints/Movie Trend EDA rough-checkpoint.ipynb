```python
# import all necessary libraries. Update this cell as work progresses
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Get dataset and merge

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")

merged_data = movies.merge(credits, left_on="id", right_on="movie_id", how="inner")
merged_data.set_index('id', inplace=True)
```

```python
# pick needed columns in merged_data
merged_data = merged_data[['budget', 'genres', 'keywords', 'original_title', 'overview', 'popularity', 
                 'production_companies', 'release_date', 'revenue', 'runtime', 'status', 'tagline',
                'vote_average', 'vote_count', 'cast', 'crew']]
merged_data.head(1)
```

```python
print(merged_data.shape)
```

# 2. Data Cleaning & Preparation

### 2.1 Handle missing values in budget, revenue, or runtime

```python
merged_data.isnull().sum()
```

```python
# to filter better since a movie cost/ profit should reasonably be higher than 1000
merged_data = merged_data[(merged_data['budget'] > 1000) & 
                (merged_data['revenue'] > 1000)]
merged_data.shape
```

```python
merged_data = merged_data.dropna(subset=['budget', 'revenue', 'runtime'])
merged_data.head(5)
```

```python
# merged_data.info()
merged_data.shape
```

### 2.2 Convert release_date to datetime and extract year

```python
merged_data['release_date'].head(3)
```

```python
# Convert release_date to datetime
merged_data['release_date'] = pd.to_datetime(merged_data['release_date'], errors='coerce')
```

```python
# Extract release year
merged_data['release_year'] = merged_data['release_date'].dt.year
merged_data = merged_data.dropna(subset=['release_year'])
```

```python
merged_data.head(1)
```

### 2.3 Split multiple genres into lists or counts

```python
import ast
```

```python
# Convert stringified list of dicts into real Python list
merged_data['genres'] = merged_data['genres'].apply(ast.literal_eval)
```

```python
# Extract just the genre names
merged_data['genres'] = merged_data['genres'].apply(lambda x: [d['name'] for d in x] if isinstance(x, list) else [])
```

```python
merged_data.head(1)
```

```python
merged_data['genre_count'] = merged_data['genres'].apply(len)
```

```python
merged_data.head(1)
```

### 2.4 Create new columns:
### profit = revenue – budget
### ROI = revenue / budget

```python
# profit
merged_data['profit']  = merged_data['revenue'] - merged_data['budget']

# ROI
merged_data['ROI']  = merged_data['revenue'] / merged_data['budget'].replace(0, pd.NA)
```

```python
merged_data.head(1)
```

# 3. Exploratory Data Analysis (EDA)

## 3.1 Descriptive stats: Average Budget, Revenue, Profit, ROI and runtime

```python
# Basic summary
summary = merged_data[['budget','revenue','profit','ROI','runtime']].agg(['count','mean','median','std','min','max'])
summary
```

```python
# Compare median and mean
print("Median budget:", round(merged_data['budget'].median()))
print("Median revenue:", round(merged_data['revenue'].median()))

print("\nAverage budget:", round(merged_data['budget'].mean()))
print("Average revenue:", round(merged_data['revenue'].mean()))
```

## 3.2 Genre analysis

```python
# Count how many movies per genre
# .explode() turns each list into multiple rows so that one genre per row is produced
# A movie that had ['Action', 'Adventure'] becomes 2 rows: one with 'Action' and one with 'Adventure'

genre_counts = merged_data['genres'].explode().value_counts()
genre_counts.head(15)
```

### 3.2.1 Which genres earn the most revenue?

```python
# Average revenue & profit per genre
genre_revenue = merged_data.assign(genre = merged_data['genres']).explode('genres').groupby('genres').agg(
    count=('original_title','count'),
    average_revenue=('revenue','mean'),
    median_revenue=('revenue','median'),
    average_profit=('profit','mean'),
    median_profit=('profit','median')
).sort_values('count', ascending=False)

genre_revenue.head(15)
```

### 3.2.2 Which are most common?

```python
# 15 most common genres
genre_counts.head(15)
```

### 3.2.3 Visualization of most common genres and Avg revenue per genre

```python
# Most common genres (bar)

plt.figure(figsize=(10,5))
genre_counts.head(15).plot(kind='bar')
plt.title("Top 12 Genres by Count")
plt.ylabel("Number of movies")
plt.show()
```

```python
# Average revenue per genre (top genres only)
plt.figure(figsize=(10,5))
genre_revenue.sort_values('median_revenue', ascending=False).head(12)['median_revenue'].plot(kind='bar')
plt.title("Top 12 Genres by Median Revenue")
plt.ylabel(f"Median revenue in dollars\n(Hundred millions)")
plt.show()
```

## 3.3 Time trends

### 3.3.0 Movies per year

```python
plt.figure(figsize = (18,10))
movies_per_year = merged_data.groupby('release_year').size()
movies_per_year.plot(kind='line', title='Movies per Year over Time')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Make sure release_year is numeric
merged_data['release_year'] = pd.to_numeric(merged_data['release_year'], errors='coerce')

# Create 5-year bins
bins = range(int(merged_data['release_year'].min()), int(merged_data['release_year'].max()) + 5, 5)
labels = [f"{b}-{b+4}" for b in bins[:-1]]

merged_data['year_bin'] = pd.cut(merged_data['release_year'], bins=bins, labels=labels, right=False)

# Count movies per 5-year bin
movies_per_bin = merged_data['year_bin'].value_counts().sort_index()

# Plot
plt.figure(figsize=(14, 5))
movies_per_bin.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Movies Released per 5-Year Interval')
plt.xlabel('5-Year Intervals')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 3.3.1 Determination of runtime variation over the years

```python
plt.figure(figsize = (12,3))
runtime_trend = merged_data.groupby('release_year')['runtime'].mean()
runtime_trend.plot(kind='line', title='Average Movie Runtime Over Time')
plt.xlabel('release year')
plt.ylabel('Average Runtime (minutes)')
```

### 3.3.2 Determination of profits variation over the years
Median is optimal for profit trends 

Reason:
Movie profits are extremely skewed. A few blockbuster hits like Avengers or Avatar make billions, while most movies barely break even or lose money.
These huge outliers pull the mean up, making it look like movies are more profitable on average than they really are.
The median gives a fairer idea of what a “typical” movie earns.

For example:
Profits = [1M, 2M, 3M, 4M, 900M]

Mean = 182M (does not reflect the true values of profits given how small the first 4 values are compared to 182M and how large 900M is compared to the mean), Median = 3M (more realistic)

```python
plt.figure(figsize = (12,3))
profits_trend = merged_data.groupby('release_year')['profit'].median()
profits_trend.plot(kind='line', title='Profits Variation Over Time')
plt.xlabel('release year')
plt.ylabel(f'Median Profits in dollars\n(hundred million)')
```

### 3.3.3 Determination of budget variation over the years
Median used for the same reason as Profits. Budgets vary a lot. Indie films may cost 500k while blockbusters cost 200M+. 

The median helps me understand what most movies cost to make, without being skewed by the few very expensive ones

```python
plt.figure(figsize = (12,3))
budget_trend = merged_data.groupby('release_year')['budget'].median()
budget_trend.plot(kind='line', title='Budget Variation Over Time')
plt.xlabel('release year')
plt.ylabel(f'Budget in dollars\n(ten million)')
```

### 3.3.4 Determination of revenue variation over the years

#### 3.3.4.1 Revenue typical appearance
Median → shows what a typical movie earns (ignores crazy blockbusters)

```python
plt.figure(figsize = (12,3))
revenue_typical_performance = merged_data.groupby('release_year')['revenue'].median()
revenue_typical_performance.plot(kind='line', title='Revenue Typical Performance Over Time')
plt.xlabel('release year')
plt.ylabel(f'Revenue in dollars\n(hundred million)')
```

#### 3.3.4.2 Industry Total Trends/ Economic Growth
Mean → shows overall market growth (includes blockbusters)

```python
plt.figure(figsize = (12,3))
revenue_total_scale = merged_data.groupby('release_year')['revenue'].mean()
revenue_total_scale.plot(kind='line', title='Revenue: Industry Total Trends/ Economic Growth Over Time')
plt.xlabel('release year')
plt.ylabel(f'Revenue in dollars\n(hundred million)')
```

```python
# Movies per year
movies_per_year = merged_data.groupby('release_year').size()

# Median runtime per year
runtime_by_year = merged_data.groupby('release_year')['runtime'].median()

# Median profit by year
profit_by_year = merged_data.groupby('release_year')['profit'].median()

# Median budget by year
budget_by_year = merged_data.groupby('release_year')['budget'].median()

# Plot
fig, ax = plt.subplots(4,1, figsize=(12,14))
movies_per_year.plot(ax=ax[0], title='Movies per Year')
runtime_by_year.plot(ax=ax[1], title='Median Runtime per Year')
profit_by_year.plot(ax=ax[2], title='Median Profit per Year')
budget_by_year.plot(ax=ax[3], title='Median Budget per Year')

for a in ax: a.set_xlabel('Year')
plt.tight_layout()
plt.show()
```

### 3.4 Budget vs Revenue: scatterplot, correlation

```python
# Scatterplot of Budget vs Revenue
plt.figure(figsize=(8, 5))
sns.scatterplot(data=merged_data, x='budget', y='revenue', alpha=0.6)

plt.title('Budget vs Revenue')
plt.xlabel('Budget (in $)')
plt.ylabel('Revenue (in $)')
plt.xscale('log')  # optional - budgets vary widely
plt.yscale('log')  # makes visualization clearer
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Calculate and interpret correlation
correlation = merged_data['budget'].corr(merged_data['revenue'])

print(f"Correlation between Budget and Revenue: {correlation:.2f}")

if correlation < 0.1:
    print("→ Very weak or no relationship. Other factors like genre, marketing, or cast may play a bigger role.")
elif correlation < 0.4:
    print("→ Weak relationship. Higher budgets don’t necessarily mean higher revenue.")
elif correlation < 0.6:
    print("→ Moderate relationship. Budget has some influence, but not always decisive.")
elif correlation < 0.8:
    print("→ Strong positive relationship. Bigger budgets generally earn more.")
else:
    print("→ Very strong relationship. High-budget movies almost always bring in high revenue.")
```

```python
# Select only numeric columns
numeric_cols = ['budget', 'revenue', 'profit', 'runtime', 'vote_average', 'vote_count', 'popularity']

# Compute correlation matrix
corr_matrix = merged_data[numeric_cols].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap of Movie Features", fontsize=14, pad=12)
plt.show()
```

### 3.5 Ratings vs Revenue: do higher-rated movies always make more money?

```python
plt.figure(figsize=(8, 5))

sns.scatterplot(
    data=merged_data,
    x='vote_average',
    y='revenue',
    alpha=0.6
)

plt.title('Ratings vs Revenue')
plt.xlabel('Average Rating (vote_average)')
plt.ylabel('Revenue ($)')
plt.yscale('log')  # use log scale since revenues vary widely
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Calculate correlation
correlation = merged_data['vote_average'].corr(merged_data['revenue'])
print(f"Correlation between Ratings and Revenue: {correlation:.2f}")

if correlation < 0.1:
    print("→ No real relationship. High ratings don't always mean high revenue.")
elif correlation < 0.4:
    print("→ Weak relationship. Critics may love a movie that doesn’t earn much, and vice versa.")
elif correlation < 0.6:
    print("→ Moderate relationship. Better ratings can help but don’t guarantee success.")
else:
    print("→ Strong positive relationship. Highly-rated movies tend to make more money.")
```

```python
rating_revenue = merged_data.groupby('vote_average')['revenue'].mean()

plt.figure(figsize=(16, 5))
rating_revenue.plot(kind='bar', color='teal', edgecolor='black')

plt.title('Average Revenue by Movie Rating', fontsize=14)
plt.xlabel('Average Rating', fontsize=12)
plt.ylabel('Mean Revenue ($)', fontsize=12)

plt.yscale('log')  # keep log scale for better visibility
# plt.xticks(rotation=15, fontsize=10)  # rotate and align labels
plt.xticks(ticks=range(0, len(rating_revenue), 2), rotation=15)
plt.tight_layout()
plt.show()
```

### 3.6 Top 10 directors/actors with the most successful films

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

# Extract directors
credits['crew'] = credits['crew'].apply(ast.literal_eval)
credits['cast'] = credits['cast'].apply(ast.literal_eval)

# Get directors
credits['director'] = credits['crew'].apply(
    lambda crew: next((member['name'] for member in crew if member['job'] == 'Director'), None)
)

# Merge with movies dataset
merged_data = movies.merge(credits, left_on='id', right_on='movie_id')

# Top 10 directors by total revenue
top_directors = (
    merged_data.groupby('director')['revenue']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

# Plot directors
plt.figure(figsize=(10,5))
top_directors.plot(kind='barh', color='teal', edgecolor='black')
plt.title('Top 10 Directors by Total Box Office Revenue')
plt.xlabel('Total Revenue (in billion $)')
plt.ylabel('Director')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()


# --- Now Top Actors ---
# Get first 3 actors per movie for relevance
merged_data['top_actors'] = merged_data['cast'].apply(lambda x: [actor['name'] for actor in x[:3]])

# Explode actor list to individual rows
actors_df = merged_data.explode('top_actors')

# Top 10 actors by total revenue
top_actors = (
    actors_df.groupby('top_actors')['revenue']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

# Plot actors
plt.figure(figsize=(10,5))
top_actors.plot(kind='barh', color='orange', edgecolor='black')
plt.title('Top 10 Actors by Total Box Office Revenue')
plt.xlabel('Total Revenue (in billion $)')
plt.ylabel('Actor')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

### 4.2 Bar charts (average ROI per genre)

```python
import ast

# Convert stringified JSON lists to actual Python objects
merged_data['genres'] = merged_data['genres'].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) else x
)

# Extract just the genre names
merged_data['genres'] = merged_data['genres'].apply(
    lambda x: [d['name'] for d in x] if isinstance(x, list) else []
)
```

```python
merged_data['budget'] = pd.to_numeric(merged_data['budget'], errors='coerce')
merged_data['revenue'] = pd.to_numeric(merged_data['revenue'], errors='coerce')
merged_data.dropna(subset=['budget', 'revenue'], inplace=True)
merged_data = merged_data[merged_data['budget'] > 0]

# Calculate ROI
merged_data['ROI'] = (merged_data['revenue'] - merged_data['budget']) / merged_data['budget']

# Explode genres and compute mean ROI
genre_roi = (
    merged_data
    .explode('genres')
    .groupby('genres')['ROI']
    .mean()
    .sort_values(ascending=False)
)

# Plot top 10
plt.figure(figsize=(10, 5))
genre_roi.head(10).plot(kind='bar', color='teal')
plt.title('Top 10 Genres by Average ROI')
plt.xlabel('Genre')
plt.ylabel('Average ROI')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 4.3 Line chart (movie revenue trends across years)

```python
# Ensure release_date is parsed as datetime
merged_data['release_date'] = pd.to_datetime(merged_data['release_date'], errors='coerce')

# Extract year from release_date
merged_data['release_year'] = merged_data['release_date'].dt.year

merged_data = merged_data.dropna(subset=['release_year', 'revenue'])
merged_data['release_year'] = merged_data['release_year'].astype(int)
merged_data['revenue'] = pd.to_numeric(merged_data['revenue'], errors='coerce')
```

```python
revenue_trend = merged_data[merged_data['revenue'] > 0].groupby('release_year')['revenue'].median()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(revenue_trend.index, revenue_trend.values, color='teal', linewidth=2)
plt.title('Median Movie Revenue Over Time')
plt.xlabel('Release Year')
plt.ylabel('Median Revenue (USD)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.plot(revenue_trend.rolling(5).mean(), color='darkorange', label='5-Year Rolling Average')
plt.legend()
```

## 5. Insights & Storytelling

This section summarizes key findings from the exploratory data analysis (EDA) of our movie dataset.  
Each insight answers a guiding question based on the visualizations and statistics above.

---

### 1. Which genres are consistently profitable?
- **Observation:**  
    Animation, Adventure, Fantasy, Family, Sci-Fi, and Action all have average profits of over a hundred million dollars. These genres consistently top the profitability charts.
- **Insight:**  
  These genres share a common thread: they are high-concept, visually driven, and have immense appeal across global and family audiences. They are the bedrock of blockbuster cinema and major franchises, leveraging spectacle and broad storytelling to guarantee strong box office returns. The high budgets often associated with them (as seen in the strong budget-revenue correlation) are justified by their proven ability to become major financial successes.

---

### 2. Are bigger budgets always worth it?
- **Observation:**  
  The correlation between **budget** and **revenue** is `0.70` 
  The scatter plot "Budget vs Revenue" confirms this, showing a clear upward trend where higher budgets generally lead to higher revenues.
- **Insight:**  
  While a hefty budget is a powerful engine for revenue, it doesn't automatically translate to profit. The "Profits Variation Over Time" chart shows that median profits, while generally positive, have significant volatility and haven't grown at the same steep rate as budgets (from "Budget Variation Over Time"). This indicates that bigger budgets increase the potential for revenue but also the financial risk. Many low-to-mid-budget films achieve superior Return on Investment (ROI) by leveraging critical acclaim or niche audience appeal, as suggested by the moderate correlation between vote_average and revenue (0.19).

---

### 3. How has Hollywood changed over the last 20 years?
- **Observation:**  
  The "Movies per Year over Time" chart shows an explosive growth in the number of films produced, particularly from the 1990s onwards. "Average Movie Runtime Over Time" reveals that films have gotten longer, increasing from around 100 minutes in the 1920s to nearly 120 minutes today. Meanwhile, "Median Profits" have trended upwards over the long term but with considerable peaks and valleys, reflecting the high-risk nature of the business.
- **Insight:**  
  The industry has shifted towards a volume-driven model, with a massive increase in output likely fueled by the rise of digital production and distribution channels like streaming services. The increasing runtimes suggest a preference for epic, immersive storytelling in cinemas, perhaps to differentiate from shorter-form content at home. The volatile but growing profits highlight that while the potential reward is greater than ever, the market is also more competitive and unpredictable.

---

### 4. Who are the most bankable actors/directors?
- **Observation:**  
  The most bankable actors by total box office revenue are Tom Cruise, Robert Downey Jr., Tom Hanks, Johnny Depp, and Ian McKellen. The most bankable directors are Steven Spielberg, Peter Jackson, James Cameron, Michael Bay, and Christopher Nolan.
- **Insight:**  
  Star power and directorial pedigree remain crucial predictors of box office success. This list is dominated by individuals associated with the high-profit genres identified earlier (Action, Adventure, Fantasy, Sci-Fi). Directors like Cameron, Jackson, and Nolan are known for their high-budget, visually spectacular films, which aligns perfectly with the strong budget-revenue correlation. Actors like Downey Jr. and McKellen are intrinsically linked to the highest-grossing film franchises in history, demonstrating the immense value of a recognizable and beloved face in a leading role.

---

### 📊 5. Summary & Takeaways
- Genre is Destiny: For consistent profitability, Animation, Adventure, and Sci-Fi are the gold standards. They transform high budgets into even higher revenues.

- Budget is a Lever, Not a Guarantee: A strong budget (correlation: 0.70) is the best predictor of high revenue, but it doesn't ensure high profits. Strategic spending is key.

- The Modern Film Landscape is Voluminous and Long-Form: The industry now produces more—and longer—films than ever before, adapting to new platforms and audience expectations.

- The Human Element Matters: Despite data and algorithms, the track record of A-list actors and visionary directors continues to be a powerful force in driving box office success, especially within the most profitable genres.
