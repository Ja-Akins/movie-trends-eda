# 1. Imports and Setup

In this section, I imported the key Python libraries for data analysis and visualization.  
Pandas handles data manipulation, NumPy supports numerical operations, Matplotlib and Seaborn are used for visualizations, and `ast` helps safely parse stringified Python objects.

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast

# Set visualization style
plt.style.use("seaborn-v0_8")
sns.set_palette("husl")

# Improve DataFrame display
pd.set_option("display.max_columns", None)
pd.set_option("display.float_format", "{:,.2f}".format)
```

# 2. Data Loading and Merging

I loaded two TMDb datasets — one containing movie information and the other containing credits (cast and crew).  
Then I merged both datasets on their common movie ID to create a single, unified DataFrame for analysis.

```python
# Load datasets
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")

# Merge datasets
merged_data = movies.merge(credits, left_on="id", right_on="movie_id", how="inner")
merged_data.set_index('id', inplace=True)

# Select relevant columns
columns_needed = [
    'budget', 'genres', 'keywords', 'original_title', 'overview', 
    'popularity', 'production_companies', 'release_date', 'revenue', 
    'runtime', 'status', 'tagline', 'vote_average', 'vote_count', 
    'cast', 'crew'
]
merged_data = merged_data[columns_needed]

print(f"Initial dataset shape: {merged_data.shape}")
```

# 3. Data Cleaning and Preparation

The merged dataset contains inconsistencies and missing values that need to be cleaned before analysis.  
I removed unrealistic values (like zero budgets), parse JSON-like text columns into lists, and extract useful features such as `release_year` and `genre_count`.  
I also created financial metrics like **profit** and **ROI (Return on Investment)** for deeper insights.

```python
def clean_and_prepare_data(df):
    
    # Create a copy to avoid modifying original
    df_clean = df.copy()
    
    # Filter out unrealistic budget/revenue values
    df_clean = df_clean[(df_clean['budget'] > 1000) & (df_clean['revenue'] > 1000)]
    
    # Handle missing values in critical columns
    df_clean = df_clean.dropna(subset=['budget', 'revenue', 'runtime'])
    
    # Convert release_date to datetime and extract year
    df_clean['release_date'] = pd.to_datetime(df_clean['release_date'], errors='coerce')
    df_clean['release_year'] = df_clean['release_date'].dt.year
    df_clean = df_clean.dropna(subset=['release_year'])
    
    # Process genres from string to list
    df_clean['genres'] = df_clean['genres'].apply(ast.literal_eval)
    df_clean['genres'] = df_clean['genres'].apply(
        lambda x: [d['name'] for d in x] if isinstance(x, list) else []
    )
    df_clean['genre_count'] = df_clean['genres'].apply(len)
    
    # Create financial metrics
    df_clean['profit'] = df_clean['revenue'] - df_clean['budget']
    df_clean['ROI'] = df_clean['revenue'] / df_clean['budget'].replace(0, pd.NA)
    
    return df_clean

# Apply cleaning function
cleaned_data = clean_and_prepare_data(merged_data)
print(f"Cleaned dataset shape: {cleaned_data.shape}")
print(f"Missing values:\n{cleaned_data.isnull().sum()}")
```

# 4. Exploratory Data Analysis (EDA)

EDA helps us understand the structure, trends, and relationships in the data.  
I started by exploring the distribution of budgets, revenues, profits, and runtimes to gain initial insights.

## 4.1 Financial Summary

We summarize the main financial variables — budget, revenue, profit, and ROI — using descriptive statistics.  
This gives a quick overview of central tendencies and ranges, helping us spot outliers and understand the data scale.

```python
def financial_summary(df):
    """Generate descriptive statistics for financial columns."""
    cols = ["budget", "revenue", "profit", "ROI", "runtime"]
    summary = df[cols].describe().T  # Transposed for easier reading
    
    # Format large numbers as currency
    for col in ["budget", "revenue", "profit"]:
        summary.loc[col, "mean"] = f"${df[col].mean():,.0f}"
        summary.loc[col, "50%"] = f"${df[col].median():,.0f}"
    
    return summary

display(financial_summary(cleaned_data))
```

## 4.2 Genre Analysis

Here, I explored which genres appear most frequently across all movies.  
By visualizing the top genres, I identified what types of films dominate the dataset and possibly link genres to profitability later.

```python
# Most common genres
genre_counts = (
    cleaned_data.explode("genres")["genres"]
    .value_counts()
    .head(10)
)

plt.figure(figsize=(8,5))
sns.barplot(x=genre_counts.values, y=genre_counts.index)
plt.title("Top 10 Movie Genres")
plt.xlabel("Number of Movies")
plt.ylabel("Genre")
plt.show()
```

```python
# Sort movies by profit
top_profitable = cleaned_data.sort_values(by="profit", ascending=False).head(10)

plt.figure(figsize=(9,5))
sns.barplot(
    data=top_profitable, 
    y="original_title", 
    x="profit", 
    palette="crest"
)
plt.title("Top 10 Most Profitable Movies")
plt.xlabel("Profit ($)")
plt.ylabel("Movie Title")
plt.show()

# Display table for clarity
top_profitable[["original_title", "budget", "revenue", "profit", "ROI", "release_year"]]
```

```python
# Average ROI per release year
roi_trend = cleaned_data.groupby("release_year")["ROI"].mean().dropna()

plt.figure(figsize=(9,5))
roi_trend.plot()
plt.title("Average ROI Over the Years")
plt.xlabel("Release Year")
plt.ylabel("Average ROI")
plt.grid(True)
plt.show()
```

# 5. Correlation and Heatmap Analysis

This section explores how numerical variables (budget, revenue, profit, ROI, etc.) relate to each other.  
The correlation matrix and heatmap visually reveal which variables move together — for example, whether higher budgets tend to generate higher revenues.

```python
# Correlation matrix
numeric_cols = ["budget", "revenue", "profit", "ROI", "runtime", "vote_average", "vote_count", "popularity"]
corr_matrix = cleaned_data[numeric_cols].corr()

# Plot correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Heatmap of Key Movie Metrics")
plt.show()
```
